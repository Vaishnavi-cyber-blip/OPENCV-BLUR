```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('taj.jpg')
average = cv2.blur(img, (5, 5))
cv2_imshow(average)
```

```python
GAUSS = cv2.GaussianBlur(img, (5, 5), 0)
cv2_imshow(GAUSS)
```
